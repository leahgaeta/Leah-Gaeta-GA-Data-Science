### Intro: Write a problem Statement/ Specific Aim for this project

Answer: Determine the association between various admission requirements and acceptance into the UCLA graduate school.  The admission requirements include GPA, GRE scores, and undergraduate school ranking, and these variables collected during one admissions cycle will be tested to predict admission into the graduate school.

### Dataset:  Write up a description of your data and any cleaning that was completed

Answer: GRE score, undergraduate GPA, and undergraduate school ranking (1 = highest, 4 = lowest) are the covariates used to predict admission into the UCLA graduate school.  In addition, outliers were tested by visualizing the data with box plots & IQR.  Null, or unknown, values were identified & dropped from the data set before analysis was done.

### Methods: Write up the methods used in your analysis

Answer: After cleaning the data, plotting the distributions of each predictor variable using histograms, and testing for outliers, the get_dummies method in pandas was used for assigning rank of the undergraduate school, since prestige is a categorical variable.  A prestige of 1 was assigned as the reference value to prevent multicolinearity. Statsmodels was then used to run a logistic regression.  In addition, odds ratios were calculated to determine the odds of being admitted based on the prestige of the undergraduate school, gpa, and gre score (95% Confidence Interval).  After creating the model, predictions for admission were made using the cartesian and linspace functions to recreate a new dataset with various possible combinations of the predictor variables.

### Results: Write up your results

Answer: The odds of admission increase with higher gpa and gre scores.  For example, a higher gpa does not guarantee admission but it improves the odds of admission when GPA is closer to 4.0.  One can also see that the probability of admission decreases with decreasing prestige of undergraduate school and that probability of admission increases significantly with higher gpa and gre scores especially when coming from a Prestige 1 school, as seen by the plots below.
